#**NORMAL** **DISTRIBUTION**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
data = np.random.normal(50, 10, 1000)

# Plot histogram
plt.hist(data, bins=30)
plt.title("Normal Distribution Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

#**MEAN,MEDIAN,VARIANCE,STANDARD DEVIATION**

```python
mean = np.mean(data)
median = np.median(data)
variance = np.var(data)
std_dev = np.std(data)

print("Mean:", mean)
print("Median:", median)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

# **PROBABLITY**

```python
# A fair coin has two outcomes: Head (H) or Tail (T).
# The probability of each outcome is P(H)=0.5 and P(T)=0.5.
# By simulating multiple tosses, we can empirically estimate these probabilities.
toss = np.random.choice(['H', 'T'], 1000)

heads = np.count_nonzero(toss == 'H')
tails = np.count_nonzero(toss == 'T')

print("Heads:", heads)
print("Tails:", tails)
print("Probability of Head:", heads/1000)
print("Probability of Tail:", tails/1000)
```
